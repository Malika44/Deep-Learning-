```python
import keras
print(keras.__version__)
import tensorflow as tf
print(tf.__version__)
```

```python
X = tf.random.uniform((100, 6))
y = (X+5)*0.3
```

```python
X.shape, y.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(name= "InputLayer", shape =[6]),
    tf.keras.layers.Dense(100, name= "Hidden_layer_1"),
    tf.keras.layers.Dense(64, name ="Hidden_layer_2"),
    tf.keras.layers.Dense(32, name = "Hidden_layer_3"),
    tf.keras.layers.Dense(16, name = "Hidden_layer_4"),
    tf.keras.layers.Dense(1, name = "Output_layer")

])
```

```python

inputs = tf.keras.layers.Input(name="InputLayer", shape=[6])

L1= tf.keras.layers.Dense(100, name="Hidden_layer_1")
L2= tf.keras.layers.Dense(64, name="Hidden_layer_2")
L3= tf.keras.layers.Dense(32, name="Hidden_layer_3")
L4 = tf.keras.layers.Dense(16, name="Hidden_layer_4")

outputs = tf.keras.layers.Dense(1, name="Output_layer")



model =\
tf.keras.models.Sequential([
    inputs,
    L1,
    L2,
    L3,
    L4,
    outputs
])
```

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_layer_names=True, show_shapes=True)
```

```python
model.compile(loss = "MAE", optimizer= "Adam", metrics= ["MSE"])
```

```python
model.fit(X, y, epochs= 100)
```
